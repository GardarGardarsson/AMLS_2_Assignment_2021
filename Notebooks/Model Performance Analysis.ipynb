```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme()

# Navigate up one directory as notebook is in ./Notebooks directory
os.chdir('..')
```

```python
train_history = pd.read_pickle('./logs/SRCNN-Losses/train_history.pkl')
val_history   = pd.read_pickle('./logs/SRCNN-Losses/val_history.pkl')
```

```python
th_per_epoch = train_history.groupby(train_history.index // 200).mean()
```

```python
th_per_epoch['train_loss']
```

```python
fig = plt.figure(figsize=(13,5),dpi=500)

plt.plot(th_per_epoch['train_loss'])
plt.plot(val_history['val_loss'])

plt.suptitle('SRCNN (x2) Learning Curves')
plt.title(r'$m=4$ $\alpha=\operatorname{adaptive}$')
plt.legend(['Training Loss (MSE)','Validation Loss (MSE)'])

plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```

```python
fig = plt.figure(figsize=(13,5),dpi=500)

plt.plot(th_per_epoch['train_psnr'])
plt.plot(val_history['val_psnr'])

plt.suptitle('SRCNN (x2) Learning Curves')
plt.title(r'$m=4$ $\alpha=\operatorname{adaptive}$')
plt.legend(['Training Loss (MSE)','Validation Loss (MSE)'])

plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```

```python
fig = plt.figure(figsize=(13,5),dpi=500)

plt.plot(th_per_epoch['train_ssim'])
plt.plot(val_history['val_ssim'])

plt.suptitle('SRCNN (x2) Learning Curves')
plt.title(r'$m=4$ $\alpha=\operatorname{adaptive}$')
plt.legend(['Training Loss (MSE)','Validation Loss (MSE)'])

plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```
